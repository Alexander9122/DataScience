```python
def obten_escenarios(nodo):
    """
    Bases = Se utilizaron los 2 posibles errores para generar un escenario posible

    Nodo = Este se separa en islas para saber en cual se encuentra la persona, siendo esta isla la que se usara para generar escenarios
    
    Creacion Escenarios = utilizando los 2 tipos de errores y el escenario posible se realizan las siguientes operaciones de conjuntos:
    1.- realiza una interseccion de la isla con el error1 de se diferente al error1 y al error2
    se genera diferencia de la isla y el error1 unido con la isla opuesta y la isla opuesta es la interseccion de la isla opuesta y el error1
    2.-aplica lo mismo que en el error 1 pero ahora la interseccion inicial es con el error2
    3.- para el caso final utiliza el escenario posible 

    De esta forma se generan todos los escenarios posibles con el nodo ingresado
    """
    error1= {'l', 'g','*'}
    error2= {'g', 'm','*'}
    posible = error1^error2
    posible.add('*')
    escenario = []
    isla1=nodo[0].copy()
    isla2=nodo[1].copy()
    nodo2=[]
    
    
    if 'p' in isla1:
        
        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla1&error1)!=error1 and (isla1&error1)!=error2:
            isla=isla1-error1
            isla2=isla2|isla
            isla1=isla1&error1
            nodo2=[isla1,isla2]

            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)

        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla1&error2)!=error1 and (isla1&error2)!=error2:
            isla=isla1-error2
            isla2=isla2|isla
            isla1=isla1&error2
            nodo2=[isla1,isla2]
            
            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)
        
        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla1&posible)!=error1 and (isla1&posible)!=error2:
            isla=isla1-posible
            isla2=isla2|isla
            isla1=isla1&posible
            nodo2=[isla1,isla2]
            
            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)
    else:        

        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla2&error1)!=error1 and (isla2&error1)!=error2:
            isla=isla2-error1
            isla1 = isla1|isla 
            isla2=isla2&error1
            nodo2=[isla1,isla2]

            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)
                
        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla2&error2)!=error1 and (isla2&error2)!=error2:
            isla=isla2-error2
            isla1 = isla1|isla 
            isla2=isla2&error2
            nodo2=[isla1,isla2]

            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)
                
        isla1=nodo[0].copy()
        isla1=set(isla1)
        isla2=nodo[1].copy()
        isla2=set(isla2)
        if (isla2&posible)!=error1 and (isla2&posible)!=error2:
            isla=isla2-posible
            isla1 = isla1|isla 
            isla2=isla2&posible
            nodo2=[isla1,isla2]

            if nodo2 not in escenario and nodo2 not in memoria:
                escenario.append(nodo2)
    
    if nodo in memoria_resultado:
        if escenario[0] not in memoria_resultado: 
            memoria_resultado.append(escenario[0])
    
    return escenario    
            
```

```python
#BFS
#Se declara el estado inicial y el final
estado_inicial = [{'m', 'l', 'g','p','*'}, {'*'}]
estado_final = [{'*'},{'p', 'm', 'l', 'g','*'}]
memoria = []
cola_bfs = [estado_inicial]
memoria_resultado = [estado_inicial]

while cola_bfs != []:  
    nodo = cola_bfs.pop(0)
    if nodo == estado_final: 
        memoria.append(nodo)
        break
    else:
        if nodo not in memoria:
            memoria.append(nodo)
        escenarios = obten_escenarios(nodo)
        cola_bfs.extend(escenarios)
solucion = recupera_solucion(memoria)
```
